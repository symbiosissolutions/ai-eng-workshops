```python
from google.colab import userdata
groq_api=userdata.get('GROQ_API')
```

```python
from langchain_groq import ChatGroq
model=ChatGroq(
    model="llama3-8b-8192",
    api_key=groq_api,
    temperature=0.5
)
```

```python
from crewai import Agent,Task, Crew, Process
```

```python
researcher=Agent(
    name="researcher",
    role="Gathering information on AI eductaion trends",
    goal="To collect relevant data and insights on how AI is being used in education",
    backstory="A curious individual who has spent years reserching the impact of AI on education",
    verbose=True,
    allow_delegation=False,
    llm=model

)
```

```python
writer=Agent(
    name="Writer",
    role="Writing a report on collected research",
    goal="To compile the research into an informative report",
    backstory="A writer with expericence in turning complex information into easy to understand documents.",
    verbose=True,
    allow_delegation=True,
    llm=model
)
```

```python
task_research=Task(
    description="Research on current trends in AI usage in education.",
    agent=researcher,
    expected_output="A list of 5 current trends in AI usage in education, each with a brief explanation."
)
```

```python
task_writer=Task(
    description="Write a report based on the research findings.",
    agent=writer,
    expected_output="A 2-page summary report that explains the 5 trends in simple and clear language."
)
```

```python
crew=Crew(
    agents=[researcher,writer],
    tasks=[task_research,task_writer]
)
```

```python
result=crew.kickoff()
```
